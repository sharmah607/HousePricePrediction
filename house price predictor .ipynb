```python

"""machine learning with kdd process"""
"""stage 1- collect data"""
import pandas as pd
df=pd.read_csv("c:/mnnit_python/Housing_Modified.csv")
df 
```

```python
#check the shape of data
df.shape

```

```python
"""stage2 -selection"""
#select dependent variable(Y)
Y=df.price

#select independent variable(X)
independent=df.columns
independent=independent.delete(0)
independent
X=df[independent]
```

```python
"""stage3 preprocessing"""
#check if the database has null values
#cleaning
Y=Y.dropna()
X=X.dropna()
print("shape of Y",Y.shape)
print("shape of X",X.shape)
```

```python
#conversion
#convert categorical to numeric
#boolean categorical to numeric
import sklearn.preprocessing as pp
lb=pp.LabelBinarizer()
X.driveway=lb.fit_transform(X.driveway)
X.recroom=lb.fit_transform(X.recroom)
X.fullbase=lb.fit_transform(X.fullbase)
X.gashw=lb.fit_transform(X.gashw)
X.airco=lb.fit_transform(X.airco)
X.prefarea=lb.fit_transform(X.prefarea)
X
```

```python
#convert categorical to numeric
df1=pd.get_dummies(X.stories,prefix="Stories")
#concatenate x and df1 on column axis
X=pd.concat([X,df1],axis=1)
```

```python
"""Stage 4 transformation"""
#change the scale of multiple variables to a common scale[0 -1]
#Algorithm: min max scaling

scale = pp.MinMaxScaler()
X = scale.fit_transform(X)
X
```

```python
"""stage-5 Data miniing"""
import statsmodels.api as sm
model =sm.OLS(Y,X).fit()
#predict using model
model.predict(X)
```

```python
#calculate multicollinearity using VIF
#vif: variance inflation factor
#check correlation among independent variables
#remove the independent variables with vif
#selection of dependent & independent variables
independent =df.columns
independent =independent.delete(0)
X=df[independent]
Y=df.price
```

```python
#import outlier influence library
#copy me leekha hai
#check correlation every column to each other column
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
#select with each column for first parameter
for i in range(len(independent)):
    mvif=[vif(X[independent].values,index) for index in range(len(independent))]
    print("VIF Values:",mvif)
    #finding max vif value
    maxvif=max(mvif)
    #if the value is greater  then 10 then delete that column
    dindex=mvif.index(maxvif)
    print("index",dindex,"MaxVIF",maxvif,"Column",independent[dindex])
    if maxvif >10:
        independent = independent.delete(dindex)
print(independent)
X= df[independent]#Independent variables
```

```python
"""bulid the model using OLS"""
import statsmodels.api as sm
# Y=m1X1 + m2X2 + ......+mnXn
#Y=denpendent,X=independent, m=weights
model= sm.OLS(Y,X)
#train the model
#calculate weight using X and Y
model=model.fit()  #weights is calculated here

#summary of the model
model.summary()
```

```python
"""COMPARE ACTUAL VS PREDICTED"""
Y_pred = model.predict(X)
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(X.lotsize,Y,color="green")
plt.scatter(X.lotsize,Y_pred,color="red")
plt.xlabel("LOT size")
plt.ylabel("price")
plt.title("actual vs predicted")
```
